**Función**

---
Para que una funcion exista se necesita dos conjuntos (A,B) y estos conjuntos se pueden operar ya sea:
1. Union
2. Intersección
3. Diferencia
4. Suma
5. Complemento

Si analizamos que los conjuntos son distintos entonces no pertenecen a un conjunto mas grande que lo llamabamos omega

Es parte importante en lo que respecta a teoria de conjuntos y podemos tener este tipo de función:
$$f(x)=3x^2$$

Entonces si vemos que

A, B son conjuntos podemos decir que:

Una función f de A o B es una regla de asociacion donde a cada elemento $$x \in A,$$ se le asigna un único elemento $$y \in B$$ al que se lo denomina como imagen de x a través de f y se lo denota $$y=f(x)$$

Por tanto podemos decir que una funcion esta compuesta por:
1. Nombre de la funcion
2. Regla de Asignacion
3. Conjunto de Partida
4. Conjunto de llegada

Resumiendolo de la siguiente manera:
$$f: A→B$$
$$f: x\rightarrow y=f(x)$$
Siendo A el conjunto de partida y B el conjunto de llegada, cuando partimos de un ejemplo vemos a la variable x sin embargo x puede pertenecer a los números naturales, enteros, racionales o incluso ser matrices por tanto se necesita completar la noción de función definiendo el resto de elementos en una regla de asociación.

**Regla de Asociación**

---
En este caso la regla de asociación es de la siguiente manera
$$f: \mathbb{R} → \mathbb{R}$$
Siendo f un subconjunto que va desde los reales hasta los reales

Por tanto en el ejemplo podemos empezar a definir las variables
$$f: x → y=f(x)=3x^2$$
definiendo nuestro punto de partida y nuesto espacio de llegada de tal manera que se pueda sacar la imagen como vemos a continuación
$$f(-5)=75$$
Donde -5 es la preimagen(dominio) y 75 es la imagen de -5 (recorrido) en la funcion

Pero no necesariamente se trabaja con conjuntos numéricos, por ejemplo se puede trabajar con:

$$Ω=\{\text palabras\ castellano\}$$
Que representa al conjunto A
$$Δ=\{a,b,\ldots ,z\}$$
Que representa al conjunto B
tal que podemos aplicar una regla de asociacion que vaya desde omega hacia delta
$$I:Ω → Δ$$
donde para cada alfa elemento de omega se le asigna un I de alfa
$$\propto → I(\propto)= \text Letra\ inicial\ de\ cada\ palabra"$$
Por ejemplo si tomamos la palabra marsupial nos queda así:
$$I(\text marsupial)= m$$
Entonces podemos notar que el concepto de funcion es bastante flexible con lo que respecta a los conjuntos y este puede ser mas amplio

**Estructura en los números reales**

---
En los números reales podemos dar una estructura de orden empezando de la siguiente manera:
$$(\mathbb{R},\leq , +, ×)$$
Empezando con comparaciones luego con operaciones aritmeticas y por último operaciones algebraicas

Por ejemplo:

Empezando con la suma primero debemos definir nuestra regla de asociacion
$$\mathbb{R} \times \mathbb{R}=\{(x,y)\}$$
que va a ser el producto cartesiano
$$+: \mathbb{R}\times \mathbb{R}→\mathbb{R}$$
luego definimos que va a estar en el conjunto de los números reales
$$+:(x,y)=x+y$$
Establecemos 2 números que estan en el conjunto de los números reales llamandolos x,y por tanto la suma queda de x+y

Por otra parte vamos con la multiplicación, establecemos la regla de asignación
$$×:\mathbb{R}^2→\mathbb{R}$$
Definimos las variables
$$×:(x,y)→x(x,y)=xy$$
por tanto el resultado va a ser la multiplicacion de esas variables

Por ultimo en la comparacion, establecemos la regla de asociación
$$\mathbb{R}_\leq : \mathbb{R}^2→\{0,1\}$$
Definimos las variables y establecemos los resultados según sea x menor o igual que y
$$\mathbb{R}_\leq : (x,y)→\mathbb{R}_\leq(x,y)=1$$
$$\text Si\ x\leq y$$
$$\mathbb{R}_\leq : (x,y)→\mathbb{R}_\leq(x,y)=0$$
$$\text Si\ no\ lo\ es$$

Entonces en el ejemplo de la anterior clase vimos
$$\mathbb{R}^\mathbb{N}=\{(x_i)_{i\in \mathbb{N}} : \forall i \in \mathbb{N},x_i\in \mathbb{R}$$
Expresando esta secuencia en función utilizando los números pares de esta forma
$$
\text{P: } \mathbb{N}^* \to \mathbb{N}^* \\
n \rightarrow P(n) = 2n
$$
donde establecemos la regla de asociacion y las variables

Pero no solo se puede partir desde un solo conjunto, también se puede partir desde los naturales a los reales
$$
x: \mathbb{N} \to \mathbb{R} \\
n \rightarrow x(n) = \frac{1}{n+1}
$$
donde partimos de los naturales llegando a los reales teniendo entonces:
$$
x_0 = 1, \quad x_1 = \frac{1}{2}, \quad x_2 = \frac{1}{5}
$$
teniendo entonces
$$
\text{Una secuencia } (x_n)_{n \in \mathbb{N}} \\
\text{que es una función real que parte de } \mathbb{N}.
$$
Hay que tomar en cuenta que la regla de asociación depende del conjunto de partida por tanto en este ejemplo depende de n

Donde f(A) esta contenido en B

Por ejemplo:

Esta funcion va de los reales a los reales, se empieza con la definicion luego se va ubicando su imagen y por ultimo su preimagen
$$
\begin{aligned}
&f: \mathbb{R} \to \mathbb{R}\\
&x \rightarrow f(x) = x^2 + 1 \\
&f(\mathbb{R}) = \{ f(x) \in \mathbb{R} : x \in \mathbb{R} \} \\
&\phantom{f(\mathbb{R})} = \{ x^2 + 1 : x \in \mathbb{R} \} \\
&\phantom{f(\mathbb{R})} = \{ x \in \mathbb{R} : x \geq 1 \} \\
&f(\mathbb{R}) = [1, +\infty) \subseteq \mathbb{R}
\end{aligned}
$$
$$f^{-1}((-10,0))$$
$$\{x\in \mathbb{R} : f(x)\in (-10,0)\}$$
$$f^{-1}((-10,0))=\{x\in \mathbb{R}: x^2+1\in (-10,0)\}$$
Como podemos ver los resultados nos dan en un intervalo de rango (-10,0)
Esta operacion esta formando la parabola que va desde los reales a los reales y va a tocar con el punto 1

![Texto alternativo](https://www.geogebra.org/resource/JraZdCtn/j1VCTmAB81aWewu0/material-JraZdCtn.png)

por tanto su dominio es $$Dom=x\in \mathbb{R}$$
y su rango es $$Rang=y\in \{1,+∞\}$$

Entonces repitiendo la operacion tenemos
$$
\begin{aligned}
&f: \mathbb{R} \to \mathbb{R}\\
&x \rightarrow x^2 + 1 \\
&f(\mathbb{R}) = [1, +\infty) \\
&f^{-1}(( -10, 0)) = \{ x \in \mathbb{R} : f(x) \in (-10, 0) \}
\end{aligned}
$$
que bien se puede sacar la imagen y una imagen bien puede ser el conjunto vacio siendo en este caso la imagen de (-10,0)

Pero que pasaria si se pide la imagen de este conjunto [0,2]
$$
f^{-1}([0, 2])
$$
Entonces primero iria la descripcción definimos el espacio donde va a estar nuetra operacion cuando pertenece a [0,2]
$$
\begin{aligned}
&\{ x \in \mathbb{R} : x^2 + 1 \in [0, 2] \} \\
&= \{ x \in \mathbb{R} : 0 \leq x^2 + 1 \leq 2 \} \\
&= \{ x \in \mathbb{R} : -1 \leq x^2 \leq 1 \} \\
&= \{ x \in \mathbb{R} : -1 \leq x \leq 1 \} \\
&f^{-1}([0, 2]) = [-1, 1]
\end{aligned}
$$
Finalmente la preimagen son todos los puntos que van desde [0,2] y que corresponde al intervalo que va desde [-1,1]

La funcion seno es una función periodica que pasa por cero y va en sentido oscilatorio

&g: \mathbb{R} \to \mathbb{R} \\

$$\begin{aligned}
&g(\mathbb{R})=[-1,1]\\

&=[-1,1]
\end{aligned}
$$
Por tanto el conjunto en el que se encuentra sin(x) es [-1,1 ]

tenemos que
$$
g^{-1}([0, +\infty)) = \{ x \in \mathbb{R} : g(x) \in [0, +\infty) \} \\
= \{ x \in \mathbb{R} : \sin(x) \geq 0 \}
