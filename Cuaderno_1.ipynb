***Tablas de verdad interactivo***

---



```python
# Importar las bibliotecas necesarias
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML

# Función para generar la tabla de verdad
def generar_tabla_verdad(conector):
    # Definimos las variables
    variables = ['A', 'B']
    # Creamos la lista de todas las combinaciones posibles de valores de verdad
    combinaciones = [(0,0), (0,1), (1,0), (1,1)]

    # Generamos la tabla de verdad según el conector seleccionado
    if conector == 'AND':
        resultado = [(A and B) for A, B in combinaciones]
    elif conector == 'OR':
        resultado = [(A or B) for A, B in combinaciones]
    elif conector == 'NOT A':
        resultado = [(not A) for A, B in combinaciones]
        variables = ['A']  # Solo una variable para NOT A
        combinaciones = [(0,), (1,)]  # Solo dos combinaciones posibles
    elif conector == 'A => B':
        resultado = [(not A or B) for A, B in combinaciones]
    elif conector == 'A <=> B':
        resultado = [(A == B) for A, B in combinaciones]

    # Convertir True/False a V/F
    resultado = ['V' if val else 'F' for val in resultado]
    combinaciones = [('V' if val else 'F' for val in comb) for comb in combinaciones]

    # Creamos el DataFrame de la tabla de verdad
    if conector == 'NOT A':
        tabla = pd.DataFrame(combinaciones, columns=variables)
        tabla[conector] = resultado
    else:
        tabla = pd.DataFrame(combinaciones, columns=variables)
        tabla[conector] = resultado

    # Mostramos la tabla de verdad
    display(tabla)

    # Explicaciones según el conector
    explicaciones = {
        'AND': 'El conector AND (A y B) es verdadero solo cuando ambos operandos son verdaderos.',
        'OR': 'El conector OR (A o B) es verdadero si al menos uno de los operandos es verdadero.',
        'NOT A': 'El conector NOT (¬A) invierte el valor de verdad de A. Es verdadero si A es falso y viceversa.',
        'A => B': 'La implicación (A => B) es falsa solo cuando A es verdadero y B es falso.',
        'A <=> B': 'La doble implicación (A <=> B) es verdadera si ambos operandos tienen el mismo valor de verdad.'
    }

    # Mostramos la explicación
    display(HTML(f"<b>Explicación:</b> {explicaciones[conector]}"))

# Creamos el menú desplegable para seleccionar el conector lógico
conector_selector = widgets.Dropdown(
    options=['AND', 'OR', 'NOT A', 'A => B', 'A <=> B'],
    description='Conector:',
)

# Creamos el botón para generar la tabla de verdad
boton_generar = widgets.Button(description='Generar Tabla de Verdad')

# Función que se llama al presionar el botón
def on_button_clicked(b):
    generar_tabla_verdad(conector_selector.value)

# Conectamos el botón con la función
boton_generar.on_click(on_button_clicked)

# Mostramos el menú desplegable y el botón
display(conector_selector)
display(boton_generar)
```
