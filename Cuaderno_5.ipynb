**Función Sobreyectiva**

---

Dado los conjuntos \( A \) y \( B \), y una función \( f \) tal que:

$$
f: A \rightarrow B
$$

donde $$ f(A) \subseteq B $$ definimos la imagen de \( A \) bajo \( f \) como:

$$
f(A) = \{ f(x) \in B : x \in A \}
$$

Decimos que \( f \) es sobreyectiva (sobretodo) si y solo si:

$$
f(A) = B
$$


$$
\text{como } f(A) \subseteq B \text{ se cumple siempre, podemos decir}
$$

$$
f \text{ sobreyectiva } \Longleftrightarrow B ⊆ f(A)
$$
Pero al igualar f(A)=B se puede decir que
$$
x  "=" y
$$
o tambien se puede decir que
$$
x \leq y \land x \geq y
$$
pero si tenemos 2 conjuntos
$$
E, F \text{ conjuntos}
$$
y los tratamos de igualar, se debe tener los mismos elementos tal que
$$
E = F \Longleftrightarrow E \subseteq F \land F \subseteq E
$$
Estableciendo la diferencia entre igualdad de conjuntos e igualdad de elementos
habiendo diferentes tipos de igualdades por ejemplo: igualdad de matrices, igualdad de funciones por tanto se debe tomar en cuenta que al igualar elementos el elemento que esta a la izquierda debe ser del mismo tipo que el elemento de la derecha, ya que si son diferentes no tiene sentido la igualdad

Definiendo la funcion sobreyectiva como
$$fsobreyectiva⟺ ∀y\in B, ∃x\in A : f(x)=y$$

![texto alternativo](https://numerosracionalesblog.wordpress.com/wp-content/uploads/2015/11/los-nc3bameros-racionales-microsoft-word_18.jpg?w=563&h=419)

Entonces si establecemos lo siguiente

$$ f : \mathbb{R} \rightarrow \mathbb{R} $$

$$ x \rightarrow f(x) = x^2 + 1 $$

$$ f(\mathbb{R}) = [1, +\infty[ $$
vemos que la grafica va desde 1 a más infinito pero en este caso no toma todos los valores de y, ya que no existe un numero real que me de 0 por ejemplo, por tanto en este caso no es una funcion sobreyectiva

Pero entonces si lo establecemos de la siguiente forma
$$ f : \mathbb{R} \rightarrow [1,+∞[ $$

$$ x \rightarrow f(x) = x^2 + 1 $$

$$ f(\mathbb{R}) = [1, +\infty[ $$
En este caso llega a ser una funcion sobreyectiva ya que empezaria desde el 1 hasta el mas infinito tomando todos los valores de la imagen y

**Función Inyectiva**

---

A diferencia de la función sobreyectiva que se podia representar a través de conjuntos, la función inyectiva tiene una version menos evidente representandola de la siguiente manera:

Una función \( f \) es inyectiva (uno a uno) si y solo si:

$$
\forall x, x' \in A, x \neq x' \Rightarrow f(x) \neq f(x')
$$
Por tanto se dice que es inyectiva si para cualquier valor de x comparado con x' sus imagenes son distintas

Lo cual es lógicamente equivalente a:

$$
\forall x, x' \in A, f(x) = f(x') \Rightarrow x = x'
$$
En caso de que sus imagenes sean iguales ya no se considera Inyectiva

Si miramos el ejemplo anterior de las funciones sobreyectivas pudimos transformar una ecuación cuadrática a sobreyectiva limitando su punto de llegada por tanto si queremos transformar esa misma función a inyectiva limitaremos su conjunto de salida de esta forma:
$$ f : \mathbb{R}^+ \rightarrow \mathbb{R} $$

$$ x \rightarrow f(x) = x^2 + 1 $$

$$ f(\mathbb{R}) = [1, +\infty[ $$

$$ f_0 : \mathbb{R}^+ \rightarrow [1, +\infty[ $$

$$ x \rightarrow f_0(x) = x^2 + 1 $$

**Función Biyectiva**

---
Por otra parte podemos decir que una funcion es biyectiva si:
$$ f \text{ es biyectiva } \iff \{ f \text{ inyectiva}, f \text{ sobreyectiva} \} $$

$$ \Updownarrow $$
Por tanto una función es biyectiva si solo si
$$ f \text{ tiene una función inversa} $$
Significando que
$$ \exists g: B \to A $$
$$ \exists g: y \to g(y) $$

**Funciones compuestas**

---
Entonces si definimos E,F,G como conjuntos
$$ \text{Sean} \; E, F, G \; \text{conjuntos} $$
y definimos f,g como funciones
$$ \text{Sean} \; f, g \; \text{funciones} $$
podemos definir la función compuesta gof
$$ g \circ f: E \to G $$
$$ g \circ f: x \mapsto (g \circ f)(x) := g(f(x)) $$

De tal manera que se necesita 3 conjuntos y 2 funciones
1. Entre el conjunto 1 y 2
2. Entre el conjunto 2 y 3

Por tanto el primer espacio es el conjunto de salida y el tercer espacio es el conjunto de llegada, siendo el segundo espacio un lugar de encuentro intermedio entre ambos

Evaluando asi a la función f(x) dentro de la función g(x)

![texto alternativo](https://matemovil.com/wp-content/uploads/2018/03/funci%C3%B3n-compuesta-diagrama-de-flechas.jpg)

Ejemplo
si definimos
$$
h(x) = \sqrt{x^2 + 1}
$$
podemos decir que es una funcion compuesta notando que basta con definir una primera función
$$
f : \mathbb{R} \longrightarrow [1, +\infty[
$$
$$
f(x) = x^2 + 1
$$
y una segunda funcion
$$
g : [1, +\infty[ \longrightarrow \mathbb{R}^+
$$
$$
g(y) = \sqrt{y}
$$
Entonces las funciones estan definidas y si queremos calcular gof hacemos lo siguiente:
$$
g \circ f : \mathbb{R} \longrightarrow \mathbb{R}^+
$$
Siendo los reales el punto de partida (1er conjunto) y el punto de llegada es los reales positivos (3er conjunto)

Entonces por la definición de la función compuesta tenemos que
$$
x→(g \circ f)(x) = g(f(x)) = g(x^2 + 1) = \sqrt{x^2 + 1}
$$
Demostrando que la función h(x) es igual a gof(x) siendo una igualdad de elementos por tanto:
$$
h(x) = (g \circ f)(x) \quad \forall x \in \mathbb{R} \quad \Longleftrightarrow \quad h = g \circ f
$$

Entonces volviendo a la función biyectiva
lo podemos denotar de la siguiente manera
$$y\to g(y): ∀x\in A, gof(x)=x$$
Donde a
$$\text{g se le denota como}\ f^-1$$
Por tanto si definimos
$$ \text{Sea} \; f: A \to B \; \text{biyectiva} $$
entonces podemos decir que
$$ \forall x \in A, \; f^{-1}(f(x)) = x $$
es decir que la imagen es unica y al ser biyectiva va a tener su inversa
